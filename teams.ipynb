```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('teams.csv')

time_intervals = ['0\'-15\'', '16\'-30\'', '31\'-45\'', '46\'-60\'', '61\'-75\'', '76\'-90\'']

teams = data['Team']
goals = data.iloc[:, 1:]

fig, axs = plt.subplots(5, 4, figsize=(15, 20))

for i in range(5):
    for j in range(4):
        ax = axs[i, j]
        team_index = i * 4 + j 
        if team_index < len(teams):
            team = teams[team_index]
            team_goals = goals.iloc[team_index]
            ax.pie(team_goals, labels=time_intervals, autopct='%1.1f%%')
            ax.set_title(f'Голы команды: {team}')

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('teams.csv')

time_intervals = ['0\'-15\'', '16\'-30\'', '31\'-45\'', '46\'-60\'', '61\'-75\'', '76\'-90\'']

teams = data['Team']
goals = data.iloc[:, 1:]

championships = np.array_split(teams, len(teams) // 4)
championship_names = ['Английская Премьер Лига', 'Ла Лига', 'Серия А', 'Бундеслига', 'Лига 1']

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, championship in enumerate(championships):
    ax = axs[i // 3, i % 3]
    championship_goals = goals.loc[championship.index]
    total_goals = championship_goals.sum()
    ax.pie(total_goals, labels=time_intervals, autopct='%1.1f%%')
    ax.set_title(f'Чемпионат: {championship_names[i]}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('teams.csv')

# Создание списка временных интервалов
time_intervals = ['0\'-15\'', '16\'-30\'', '31\'-45\'', '46\'-60\'', '61\'-75\'', '76\'-90\'']

# Извлечение команд и их голов из данных
teams = data['Team']
goals = data.iloc[:, 1:]

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(30, 15))

colors = [
    '#808080', '#A9A9A9', '#C0C0C0', '#D3D3D3',  # Оттенки серого
    '#FFFF00', '#FFD700', '#FFC400', '#FFB100',  # Оттенки жёлтого
    '#00FF00', '#00D600', '#00C100', '#00AE00',  # Оттенки зелёного
    '#FF0000', '#DC0000', '#C80000', '#B40000',  # Оттенки красного
    '#0000FF', '#0000DC', '#0000C8', '#0000B4',  # Оттенки синего
]

for i, interval in enumerate(time_intervals):
    ax = axs[i//3, i%3]
    ax.bar(teams, goals[interval], color=colors)
    ax.set_ylim(0, 30)  # Установка диапазона по оси Y
    ax.set_ylabel('Количество голов')
    ax.set_title(f'Голы команд по временному интервалу: {interval}')

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('teams.csv')

time_intervals = ['0\'-15\'', '16\'-30\'', '31\'-45\'', '46\'-60\'', '61\'-75\'', '76\'-90\'']

teams = data['Team']
goals = data.iloc[:, 1:]

championships = np.array_split(teams, len(teams) // 4)
championship_names = ['Английская Премьер Лига', 'Ла Лига', 'Серия А', 'Бундеслига', 'Лига 1']

average_goals = []

for championship in championships:
    championship_goals = goals.loc[championship.index]
    average_championship_goals = championship_goals.mean()
    average_goals.append(average_championship_goals)

df_average_goals = pd.DataFrame(average_goals, columns=time_intervals, index=championship_names)

fig, ax = plt.subplots(figsize=(10, 6))
df_average_goals.plot(kind='bar', ax=ax)
ax.set_xlabel('Чемпионат', weight='bold')
ax.set_ylabel('Среднее количество голов')
ax.set_title('Среднее количество голов по временным интервалам для каждого чемпионата')
plt.xticks(rotation=0)


plt.show()
```
